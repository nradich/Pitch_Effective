Its spring training. Beat writers around the MLB are talking to players at camp about adjustments they've made to their games over the off season. For many pitchers, this means adding a new pitch to repertoire. This analysis will look to explore historically what ptiches generate outs or weak contact for pitcher and will predict which pitch may be the out pitch of choice for the 2019 season.

Goal of this project: visualize historically what pitches generate weak contact, outs, swings and misses.
Could attempt to predict what pitch will be on the rise for the 2019 season, or bring light to some correlation between the matters

Concerns: This project won't take into account pitching sequencing which is a huge factor in swing miss/

1999 is the last year with available data

Documentation for column names: https://baseballsavant.mlb.com/csv-docs
